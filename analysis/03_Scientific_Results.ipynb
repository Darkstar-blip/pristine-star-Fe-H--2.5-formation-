```python
import numpy as np
import matplotlib.pyplot as plt
import corner

# Load samples from the main pipeline
samples = np.load('../data/mcmc_samples.npy')

# Generate the Corner Plot
fig = corner.corner(samples, 
                    labels=["SFR Efficiency", "IMF Slope ($\alpha$)"],
                    color='teal',
                    show_titles=True)

plt.savefig("../plots/corner_plot.png")
plt.show()
```
