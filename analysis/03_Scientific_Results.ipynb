```python
import numpy as np
import matplotlib.pyplot as plt
import corner

# 1. Load the results from your main pipeline
# Ensure the path points to your data folder
data_path = '../data/mcmc_samples.npy'
samples = np.load(data_path)

# 2. Define the labels for your research variables
# Theta 1: Star Formation Efficiency (SFE)
# Theta 2: IMF Slope (alpha) - Salpeter is 2.35
labels = ["SFE (Formation Efficiency)", r"IMF Slope ($\alpha$)"]

# 3. Generate the Corner Plot
fig = corner.corner(
    samples, 
    labels=labels,
    truths=[0.1, 2.35],           # Overlay the "Standard Model" values
    show_titles=True,             # Display the numerical results above histograms
    title_kwargs={"fontsize": 12},
    color="midnightblue",         # Professional science aesthetic
    truth_color="crimson",        # Highlights where the theory meets your data
    quantiles=[0.16, 0.5, 0.84],  # Shows 1-sigma uncertainties
    plot_contours=True
)

# 4. Save to your plots folder for your GitHub README
plt.savefig("../plots/corner_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```
